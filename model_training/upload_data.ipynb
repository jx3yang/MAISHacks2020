```python
chunk_size = 100
sample_size = 200

for key in features.keys():
    if key not in ['name', 'timestamp']:
        df = pd.DataFrame([features['name'], features['timestamp'], features[key]])
        df = df.rename(index={key: 'value'})
        
        data = list(df.to_dict().values())[:sample_size]
        for i in range(len(data) // chunk_size - 1):
            requests.post(f"http://127.0.0.1:3500/api/add_metric/{key}/", data=json.dumps(data[i * chunk_size: (i + 1) * chunk_size]))
            
```
